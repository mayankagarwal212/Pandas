```python
import numpy as np
import pandas as pd
```

# **Creating DataFrame**
for creating dataframe you can make it by list, array, dis etc...

```python
data = {
    'Name' : ['John', 'Anna', 'Peter', 'Linda'],
    'Age' : [28,32,19,38],
    'City' : ['New York', 'Paris', 'Barlin', 'London'], 
    'Salary' : [65000, 80000, 62000, 75000]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
data_list = [
    ['John',28,'New York', 65000],
    ['Anna', 32, 'Paris', 80000],
    ['Peter', 19, 'Barlin', 75000],
    ['Linda', 48, 'Londan', 62000]
]
df2 = pd.DataFrame(data_list)
```

```python
df2
```

If you want there is a columns name so you can do that by 

```python
columns = ['Name', 'Age', 'City', 'Salary']
df2 = pd.DataFrame(data_list, columns = columns)
```

```python
df2
```

# **Selection and Indexing of Colums**

If you want to target an specific column then 

```python
df2['City']
```

I you want to target more then one column then


```python
df2[['Name', 'City', 'Salary']]
```

# **Creating a new columns**

If you want to add new columns in DataFrame then you have to provide exactly no of value that present in Frame otherwise it gives error.

```python
df2['Designation'] = ['Doctor', 'Engineering', 'Doctor', 'Engineering']
```

```python
df2
```

# **Removing Columns**
If you want to remove column in DataFrame the you have to use drop and provide axis by default it set 0 but here we want to remove designation column then we need to provide axis as 1.

```python
df2.drop('Designation', axis = 1)
```

Here is a point that drop is not delete the column in main data it show that how the data look after remove.

For permanent delete you have to use *inplace = True*

```python
df2.drop("Designation", axis = 1, inplace = True)
```

```python
df2
```

If you want to remove multiple Series then you can do it like that

```python
df2.drop(['Age', 'Salary'], axis = 1)
```

```python
df2
```

```python
df.drop(0,axis = 0)
```

# **Selecting Row**
For selecting a row you have to use loc[] with DataFrame like that

```python
df2.loc[0]
```

And for selecting multiple row then

```python
df2.loc[[0,1]]
```

# **Selecting Subsets of Row and Columns**

for selection subpart of table you have to provide first row detail in first[] and then column detail in 2nd [] like that.

```python
df.loc[[0,1]][['City','Salary']]
```

```python
df2.loc[[2,3]][['Name', 'Age']]
```

# **Condition Selection**

Example : I want to select those people whose age is more then 30

```python
df2[df2['Age'] > 30]
```

IF I inly want those people who is more then 30 and blongs to Paris

```python
df2[(df2['Age'] > 30) & (df2['City'] == 'Paris')]
```
