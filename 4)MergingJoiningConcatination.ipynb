```python
import numpy as np 
import pandas as pd 
```

# **Merging 2 DataFrames**

```python
# DataFrame 1
employee = pd.DataFrame({
    "Employee_id" : [1,2,3,4,5],
    "Name" : ['John', "Anna", "Peter", "Linda", "Bob"],
    "Department" : ["HR", "IT", "Finance", "IT", "HR"]
})

# DataFrame 2
salaries = pd.DataFrame({
    "Employee_id" : [1,2,3,6,7],
    "Salary" : [50000, 72000, 55000, 80000, 45000],
    "Bonus" : [5000, 10000, 7000, 9000, 6000]
})
```

```python
employee
```

```python
salaries
```

To merge these DataFrame we use *merge()*

```python
pd.merge(employee, salaries, on= 'Employee_id', how= 'inner')
```

```python
pd.merge(employee, salaries, on= 'Employee_id', how= 'outer')
```

```python
pd.merge(employee, salaries, on='Employee_id', how='left')
```

```python
pd.merge(employee, salaries, on= "Employee_id", how= 'right')
```

# **Concatination of 2 Dataframe**

```python
df1 = pd.DataFrame({
    'A' : ['A0', 'A1', 'A2'],
    'B' : ['B0', 'B1', 'B2'],
    'C' : ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A' : ['A3', 'A4', 'A5'],
    'B' : ['B3', 'B4', 'B5'],
    'C' : ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis = 1)
```

# **Joining 2 DataFrame**

```python
df1 = pd.DataFrame({
    'name' : ['Alice', 'Bob', "Charlie"]
}, index = [1,2,3])

df2 = pd.DataFrame({
    'score' : [85,90,75]
}, index = [2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2, how='outer')
```
