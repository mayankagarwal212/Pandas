# **Pandas Series**
A Series is a one dimensional labeled array capable of holding any data type. The axis labels are collectively called the index.

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a','b','c']
my_list = [10,20,30]
my_arr = np.array([12,3,42])
d = {1:67, 2:9, 4:8}
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list, index = labels)
```

```python
pd.Series(my_arr)
```

```python
pd.Series(d)
```
